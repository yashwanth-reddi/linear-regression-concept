```python

# original outputs 12 29 32 23
    
    
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('E:\\test.csv',nrows=4)
```

```python
train = data[['a','b']]
y     = data.c.values
y
```

```python
train
```

```python
train.shape[1],1
x=train
```

```python
w= np.zeros((train.shape[1], 1))
eta = 0.03
iterations = 15
m = train.shape[0]
w
```

```python
train.T
```

# without intercept

```python
for it in range(1000+1):
    y_pred = np.dot(train,w)
    residues = y_pred - y.reshape([train.shape[0],1])
    gv = np.dot(train.T,residues)
    w  -= eta/m * gv
    cost = np.sum((residues ** 2)) / (2 * m)
    if it%100 ==0 :
        print(it,"th iteration")
        print("y= ",y)
        print("y_pred = ",y_pred)
        print("residues = ",residues)
        print("gv =",gv)
        print("w =",w)
        print("cost ", cost,"\n\n")
```

## weights and coefficients are equal


```python
lr = LinearRegression(copy_X=True,fit_intercept=False,n_jobs=None,normalize=False)
lr.fit(train,y)
print(lr.coef_)
print(lr.intercept_)
w
```
